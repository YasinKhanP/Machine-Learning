```python
#numpy is used to perform operation on arrays
import numpy as np
#matplotlib.pyplot is used to plot the graph
import matplotlib.pyplot as plt
#sklearn.dataset is used to load the predefined dataset
from sklearn.datasets import load_diabetes
#sklearn.linear_model is used to load the linear regression model
from sklearn.linear_model import LinearRegression
#sklearn.model_selection is used to split the data into training and testing data
from sklearn.model_selection import train_test_split
#sklearn.metrics is used to calculate the accuracy of the model
from sklearn.metrics import mean_squared_error, r2_score
# Load the diabetes dataset
from sklearn import linear_model
x,y=load_diabetes(return_X_y=True)
print(x.shape)
print(y.shape)
print(x)
print(y)
# Split the data into training and testing sets
x=x[:,8].reshape(-1,1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
lr=linear_model.LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
plt.scatter(x_test,y_test,color="black",label="original data")
plt.plot(x_test,y_pred,color="blue",linewidth=3,label="regression line")
plt.xlabel("Independent Feature")
plt.ylabel("Target values")
plt.title("Simple Linear Regression")
plt.show()
```
